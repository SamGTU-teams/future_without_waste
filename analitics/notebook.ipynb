```python
import pandas as pd

data = pd.read_csv('data.csv')

data["population"] = pd.to_numeric(data["population"])

data.info()
```

```python

for typ in ["разъезд", "посёлок, станция", "хутор", "стация", "местечко"]:
    data = data[data["type"] != typ]

data_not_0 = data[data["population"]!=0]

data.info()
```

```python
data["population"].sum()
```

```python
print(data["population"].mean())
print(data_not_0["population"].mean())
```

```python
print(data["population"].median())
print(data_not_0["population"].median())
```

```python
print(data["locality"].count())
print(data_not_0["locality"].count())
```

```python
print(data[data["population"]<100]["locality"].count())
print(data_not_0[data_not_0["population"]<100]["locality"].count())
```

```python
print(data[data["population"]<1000]["locality"].count())
print(data_not_0[data_not_0["population"]<1000]["locality"].count())
```

```python
print(data.groupby("district")["population"].sum().sort_values(ascending=False))
print(data_not_0.groupby("district")["population"].sum().sort_values(ascending=False))
```

```python
print(data[data["population"]>100].groupby("district")["population"].mean().sort_values(ascending=False))
print(data_not_0[data_not_0["population"]>100].groupby("district")["population"].mean().sort_values(ascending=False))
```

```python
rang = [0, 155, 900, 5000, 200000, 300000]

for i in range(1, len(rang)):
    bf = data[data["population"]<rang[i]]
    d = bf[bf["population"]>=rang[i-1]]
    popul = d["population"]
    print(f"from {rang[i-1]} to {rang[i]}:")
    print("count:", popul.count())
    print("sum:", popul.sum())
    print("mean:", popul.mean())
    print("median:", popul.median())
    print("min:", popul.min())
    print("max:", popul.max())
    print("Top:")
    print(d.sort_values("population", ascending=False)[:10])

    print()
    
```
